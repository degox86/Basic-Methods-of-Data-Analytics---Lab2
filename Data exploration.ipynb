```python
df_MCI['occurrencedate'] = pd.to_datetime(df_MCI.occurrencedate)
```

```python
df_MCI_hotspots=df_MCI.groupby(['occurrencedate_M','occurrenceday','Division','Hood_ID','Neighbourhood'])['event_unique_id'].count().reset_index()
```
